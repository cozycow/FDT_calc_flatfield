```python
%matplotlib qt
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
import glob

from cavity_correction import *
from limb_fitting import *
from imageutils import *
from calc_flatfield import *
```

```python
folder = 'temp/'
files = sorted(glob.glob(folder + '*.fits'))
files
```

```python
images = []

for file in files[:]:
    with fits.open(file) as hdul:
        image = hdul[0].data[2]
        header = hdul[0].header
    images.append(image)
```

```python
flatfield = calc_transmittance(images)
```

```python
np.savez('transmittance.npz', flatfield=flatfield)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(flatfield, vmin=0.9, vmax=1.1)
plt.tight_layout()
```
